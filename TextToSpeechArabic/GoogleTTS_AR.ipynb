```python
from gtts import gTTS
from google.colab import files
import IPython

d = {
     "English":'en',
     "Spanish":'es',
     "Indian":'hi',
     "Arabic":'ar',
     "French":'fr',
     "Chinese":'zh',
     "Polish":'pl',
     "Ukrainian":'uk',
     "Afrikaans":'af'

}
text = "\u0647\u0630\u0627 \u0627\u0644\u0646\u0635 \u0645\u0643\u062A\u0648\u0628 \u0628\u0627\u0644\u0644\u063A\u0629 \u0627\u0644\u0639\u0631\u0628\u064A\u0629 \u0645\u0631\u062D\u0628\u0627 \u0628\u0643 " #@param {type:"string"}
Accents = 'Arabic' #@param ["English", "Spanish","Afrikaans", "Indian", "Arabic", "Chinese", "Polish", "French", "Ukrainian"]

pytts = gTTS(text, lang=d[Accents])

pytts.save('audio_generated.mp3')

IPython.display.Audio("audio_generated.mp3")
# files.download("audio_generated.mp3")
```
